# Task 1: Basic Python Programming (20 Marks)

1. Write a function `fibonacci(n)` that generates the first `n` numbers in the Fibonacci sequence and returns

```python
def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]

    fib_seq = [0, 1]
    for i in range(2, n):
        next_number = fib_seq[-1] + fib_seq[-2]
        fib_seq.append(next_number)

    return fib_seq

print(fibonacci(8))
```

2. Write a function `is_prime(num)` that checks if a given number is a prime number. Return `True` if the number

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

print(is_prime(17))
print(is_prime(20))
```

```python
def convert_sex(x):
  if x == 'male':
    return 0
  else:
    return 1

df['Sex'] = df['Sex'].apply(convert_sex)
df
```
